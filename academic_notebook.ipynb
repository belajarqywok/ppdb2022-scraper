<ul><li><h4>Determine The Capacity Of Academic Students In Each Administrative City</h4></li></ul>

```python
print("\n[ Determine The Capacity Of Academic Students In Each Administrative City ]")

city_name = [
    "Kota Jakarta Timur",
    "Kota Jakarta Pusat",
    "Kota Jakarta Barat",
    "Kota Jakarta Selatan",
    "Kota Jakarta Utara",
    "Kab. Kepulauan Seribu"
]

school_capacities = np.zeros(len(city_name))

for index_academic_profiles_values in range( len(academic_profiles_values) ) :
    
    for index_city_name in range( len(city_name) ) :
    
        if academic_profiles_values[index_academic_profiles_values][7] == city_name[index_city_name] :
            
            capacity = academic_profiles_values[index_academic_profiles_values][11]
            school_capacities[index_city_name] += capacity
            
            
for index_show_capacities in range( len(city_name) ) :
    print(f"\n{city_name[index_show_capacities]} : {school_capacities[index_show_capacities]} Students")
    
print(f"\n\nTotal : {sum(school_capacities)} Students")
```

<br>
<br>
<ul><li><h4>Determine The Percentage Of Academic Students In Each Administrative City</h4></li></ul>

```python
print("\n[ Determine The Percentage Of Academic Students In Each Administrative City ]")

city_percentage = np.zeros(len(city_name))

for index_city_percentage in range( len(school_capacities) ) :
    
    percentage = ( school_capacities[index_city_percentage] / sum(school_capacities) ) * 100
    city_percentage[index_city_percentage] += percentage
    
    print(f"\n{city_name[index_city_percentage]} : {city_percentage[index_city_percentage]} %")
    
    city_name[index_city_percentage] += f"  ( {percentage}  % )"

print(f"\n\nTotal : {sum(city_percentage)} %")
```

```python
print("\n\n\n         [ Plots Of Academic Students In Each Administrative City ( Bar, Line & Scatter Chart ) ]\n\n")

plt.barh(city_name, school_capacities, color=["#1E88E5"])
plt.plot(school_capacities, city_name, color="blue")
plt.scatter(school_capacities, city_name, color="black")
plt.show()
```

<br>
<br>
<ul><li><h4>Plots Of Academic Students In Each Administrative City ( Pie Charts )</h4></li></ul>

```python
print("\n\n\n   [  Plots Of Academic Students In Each Administrative City ( Pie Charts )  ]\n\n")

for index_plot in range( len(city_name) ) :
    print(f"\n\n                     [ {city_name[index_plot]} ] \n")
    
    explode = np.zeros(len(city_name))
    
    for index_explode in range( len(explode) ) :
        if index_plot == index_explode :
            if index_explode == 5 :
                explode[index_explode] += 1.2
            else :
                explode[index_explode] += 0.5
    
    plt.pie(school_capacities, labels = city_name, explode = explode, shadow = True, startangle = 0)
    plt.show()
    print("\n\n\n")
```

<ul><li><h4>Retrieve Datas From Each Administrative City</h4></li></ul>

```python
academic_csv_location = "datasets/smp/prestasi_akademik/participants/"

school_datas = []

for index_school_datas in os.listdir(academic_csv_location) :
    
    read_csv_students = pd.read_csv(academic_csv_location + index_school_datas, sep=",", index_col=0)
    
    read_csv_students_values = read_csv_students.values
    
    school_student_values = []
    
    for index_school_student_values in read_csv_students_values :
        school_student_values.append(
            index_school_student_values[1]
        )
        
    school_name = index_school_datas[0:len(index_school_datas)-len(".csv")]
    sum_value   = sum(school_student_values)
    max_value   = max(school_student_values)
    min_value   = min(school_student_values)
    
    
    # get school location
    
    for index_academic_profiles_values in academic_profiles_values :
        
        if index_academic_profiles_values[1] == school_name :
            
            location_school = index_academic_profiles_values[7]
            break
            
        else :
            pass
    
    school_datas.append({
        
        "school_name"     : school_name,
        "location_school" : location_school,
        "max_value"       : max_value,
        "min_value"       : min_value,
        "sum_value"       : sum_value
        
    })
    
pd.DataFrame(school_datas)
```

<ul><li><h4>Determine The Amount Of Student Scores ( V < 60, V = 60, 60 < V )</h4></li></ul>

```python
print("\n[ Determine The Amount Of Student Scores ( V < 60, V = 60, 60 < V ) ]")

mid_value = 60

equal_than   = 0
greater_than = 0
smaller_than = 0

for index_academic_amount_of_value in os.listdir(academic_csv_location) :
    
    read_csv_students = pd.read_csv(academic_csv_location + index_academic_amount_of_value, sep=",", index_col=0)
    
    read_csv_students_values = read_csv_students.values
    
    for index_school_student_values in read_csv_students_values :
        
        if index_school_student_values[1] == mid_value :
            equal_than += 1
            
        elif index_school_student_values[1] < mid_value :
            smaller_than += 1
            
        elif index_school_student_values[1] > mid_value :
            greater_than += 1
            
            
labels = [f"V < {mid_value}", f"V = {mid_value}", f"{mid_value} < V"]
amount = [smaller_than, equal_than, greater_than]

plt.barh(labels, amount, color=["#1E88E5"])
plt.plot(amount, labels, color="blue")
plt.scatter(amount, labels, color="black")
plt.plot()

print(f"\n\n             greater_than ( {mid_value} < N ) : {greater_than}")
print(f"             equal_than   ( N = {mid_value} ) : {equal_than}")
print(f"             smaller_than ( N < {mid_value} ) : {smaller_than}\n")
```

<ul><li><h4>Determine Student Rankings Across Administrative Cities</h4></li></ul>

```python
print("\n[ Determine Student Rankings Across Administrative Cities ]")

rank = []

for index_academic_ranking in os.listdir(academic_csv_location) :
    
    read_csv_students = pd.read_csv(academic_csv_location + index_academic_ranking, sep=",", index_col=0)
    
    read_csv_students_values = read_csv_students.values
    
    for index_school_student_values in read_csv_students_values :
        
        rank.append({
            "student_name" : index_school_student_values[0],
            "value"        : index_school_student_values[1]
        })
        

for student_sorting_1 in range( len(rank) ) :
    
    for student_sorting_2 in range( len(rank) - student_sorting_1 ) :
        
        if student_sorting_2 == len(rank) - 1 :
            pass
        
        elif rank[student_sorting_2]["value"] < rank[student_sorting_2 + 1]["value"] :
            rank[student_sorting_2 + 1], rank[student_sorting_2] = rank[student_sorting_2], rank[student_sorting_2 + 1]

pd.DataFrame(rank)
```

```python
search_name = ["ALADDIN NAUFAL SURAHMAN", "ALLISYA AL-ZIHAN RACHEL", "ALFIO RASHAAD KALOKO", "PASYA RAMADHAN"]

def search(name) :
    for index_rank in range(len(rank)) :

        if rank[index_rank]["student_name"] == name :

            print(f'\n\nranking : {index_rank+1} / {len(rank)} student')
            print(f'name    : {rank[index_rank]["student_name"]}')
            print(f'value   : {rank[index_rank]["value"]}')

            break

for index_search_name in search_name :
    search(index_search_name)
```

<ul><li><h4>Determine The Rank Of Students On The Path Of Academic Achievement In Each Administrative City</h4></li></ul>

```python
city_name = [
    "Kota Jakarta Timur",
    "Kota Jakarta Pusat",
    "Kota Jakarta Barat",
    "Kota Jakarta Selatan",
    "Kota Jakarta Utara",
    "Kab. Kepulauan Seribu"
]


administrative_students   =  [ [] for i in range(len(city_name)) ]


for index_school_datas in os.listdir(academic_csv_location) :
    
    read_csv_students = pd.read_csv(academic_csv_location + index_school_datas, sep=",", index_col=0)
    
    read_csv_students_values = read_csv_students.values
    school_name = index_school_datas[0:len(index_school_datas)-len(".csv")]
    
    for index_school_name in academic_profiles_values :
        
        if school_name == index_school_name[1] :
    
            for index_city_name in range( len(city_name) ) :

                if city_name[index_city_name] == index_school_name[7] :
                    
                    for index_read_csv_students_values in read_csv_students_values :
                        
                        administrative_students[index_city_name].append({
                            "student_name" : index_read_csv_students_values[0],
                            "value"        : index_read_csv_students_values[1]
                        })

                        
# sorting

for index_administrative_rank in range( len(city_name) ) :
    
    for student_sorting_1 in range( len(administrative_students[index_administrative_rank]) ) :
    
        for student_sorting_2 in range( len(administrative_students[index_administrative_rank]) - student_sorting_1 ) :
        
            if student_sorting_2 == len(administrative_students[index_administrative_rank]) - 1 :
                pass
        
            elif administrative_students[index_administrative_rank][student_sorting_2]["value"] < administrative_students[index_administrative_rank][student_sorting_2 + 1]["value"] :
                administrative_students[index_administrative_rank][student_sorting_2 + 1], administrative_students[index_administrative_rank][student_sorting_2] = administrative_students[index_administrative_rank][student_sorting_2], administrative_students[index_administrative_rank][student_sorting_2 + 1]
```

<ul><li><h4>Determine The Average Value Of The Administrative City Of The Academic Achievement Path</h4></li></ul>

```python
city_name = [
    "Kota Jakarta Timur",
    "Kota Jakarta Pusat",
    "Kota Jakarta Barat",
    "Kota Jakarta Selatan",
    "Kota Jakarta Utara",
    "Kab. Kepulauan Seribu"
]

average_academic_administrative = np.zeros( len(city_name) )

for index_average_academic_administrative in range( len(average_academic_administrative) ) :
    
    temp_value = []
    
    for index_temp_value in administrative_students[index_average_academic_administrative] :
        
        temp_value.append(index_temp_value["value"])
    
    sum_administrative_students = sum(temp_value)
    len_administrative_students = len(temp_value)
    
    average_value = sum_administrative_students/len_administrative_students
    
    average_academic_administrative[index_average_academic_administrative] += average_value
    

print("\n\n\n[ Determine The Average Value Of The Administrative City Of The Academic Achievement Path ]\n\n")

for adm in range( len(average_academic_administrative) ) :
    print(f"{city_name[adm]} : {average_academic_administrative[adm]}")
    
print(f"\nTotal : {sum(average_academic_administrative)}")
    
print("\n")
    
plt.barh(city_name, average_academic_administrative, color=["#1E88E5"])
plt.plot(average_academic_administrative, city_name, color="blue")
plt.scatter(average_academic_administrative, city_name, color="black")
plt.plot()
plt.show()

print("\n\n\n")

for index_city_pie in range( len(city_name) ) :
    
    print(f"\n\n\n\n\t\t\t\t[ {city_name[index_city_pie] } ]\n\n")
    
    percantage = (average_academic_administrative[index_city_pie]/sum(average_academic_administrative))*100
    
    explode = np.zeros(len(city_name))
    
    for index_explode in range( len(explode) ) :
        if index_city_pie == index_explode :
            explode[index_explode] += 0.5
                
    
    city_name[index_city_pie] += f"  ( { percantage } % )"

    plt.pie(average_academic_administrative, labels = city_name, explode = explode, shadow = True, startangle = 90)
    plt.show()
```
